```python
from scipy.signal import stft
import yfinance as yf
import math
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Define the ticker symbol
ticker_symbol = 'AAPL'

# Get data for the specified ticker symbol
stock_data = yf.Ticker(ticker_symbol)

# Retrieve historical data
historical_data_apple = stock_data.history(period='5y')
time_index = historical_data_apple.index


fig, axs = plt.subplots(2, figsize=(8, 6))
axs[0].plot(historical_data_apple.index, historical_data_apple['Close'], label='APPL stock price 5 years', color='blue')
axs[0].set_title('APPL Stock')
axs[0].set_xlabel('Date')
axs[0].set_ylabel('Price')
axs[0].legend()
axs[0].grid(True)
axs[0].xaxis.set_tick_params(rotation=45)





#Define the ticker symbol
ticker_symbol = 'MSFT'

# Get data for the specified ticker symbol
stock_data = yf.Ticker(ticker_symbol)

# Retrieve historical data
historical_data_msft= stock_data.history(period='5y')
time_index = historical_data_msft.index


axs[1].plot(historical_data_msft.index, historical_data_msft['Close'], label='MSFT stock price 5 years', color='red')
axs[1].set_title('MSFT stock')
axs[1].set_xlabel('Date')
axs[1].set_ylabel('Price')
axs[1].legend()
axs[1].grid(True)
axs[1].xaxis.set_tick_params(rotation=45)



plt.tight_layout()
plt.show()
```
