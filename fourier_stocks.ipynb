```python
import yfinance as yf
import math
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Define the ticker symbol
ticker_symbol = 'AAPL'

# Get data for the specified ticker symbol
stock_data = yf.Ticker(ticker_symbol)

# Retrieve historical data
historical_data = stock_data.history(period='5y')

# Plotting historical closing prices
plt.figure(figsize=(10, 6))
plt.plot(historical_data.index, historical_data['Close'], label='Close Price', color='blue')

# Adding title and labels
plt.title('Historical Close Price of {}'.format(ticker_symbol))
plt.xlabel('Date')
plt.ylabel('Close Price')

# Rotating x-axis labels for better readability
plt.xticks(rotation=45)

# Displaying legend
plt.legend()

# Show plot
plt.tight_layout()
plt.show()
```

```python
# Get data for the specified ticker symbol
stock_data = yf.Ticker(ticker_symbol)

# Retrieve historical data
historical_data = stock_data.history(period='5y')

# Extract closing prices
closing_prices = historical_data['Close'].values

# Number of data points
N = len(closing_prices)

# Time interval (assuming daily data)
dt = 1

# Compute Fourier Transform
frequencies = np.fft.fftfreq(N, dt)
fourier_transform = np.fft.fft(closing_prices)

# Plot Fourier Transform
plt.figure(figsize=(10, 6))
plt.plot(frequencies, np.abs(fourier_transform))
plt.title('Fourier Transform of Closing Prices ({})'.format(ticker_symbol))
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
# # Identify the index corresponding to the frequency you want to remove
# # For example, if you want to remove the frequency at index i, set fourier_transform[i] = 0
# # Replace index_i with the appropriate index
# index_i = 0  # Example frequency index

# # Set the magnitude of the frequency component to zero
# fourier_transform[index_i] = 0


# Set threshold for amplitude
threshold_amplitude = 5000  # Set your threshold here

# Identify frequency components above threshold amplitude and set them to zero
for i, amplitude in enumerate(np.abs(fourier_transform)):
    if amplitude > threshold_amplitude:
        fourier_transform[i] = 0

# Plot Fourier Transform
plt.figure(figsize=(10, 6))
plt.plot(frequencies, np.abs(fourier_transform))
plt.title('Fourier Transform of Closing Prices ({})'.format(ticker_symbol))
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.xlim(-0.1, 0.1)
plt.show()
```

```python
time_index = historical_data.index

modified_signal_apple = np.fft.ifft(fourier_transform)

# Plot the original closing prices
plt.figure(figsize=(12, 6))
# plt.plot(time_index, closing_prices, label='Original Closing Prices', color='blue')

# Plot the modified signal with the same time index as the original data
plt.plot(time_index, modified_signal_apple.real, label='Modified Signal', color='red')

plt.title('Original Closing Prices vs. Modified Signal')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
